```python
plt.figure(figsize=(8,5))
sns.histplot(df_clean['price'], bins=40, kde=True)
plt.title("Price Distribution of Laptops")
plt.xlabel("Price (INR)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df_clean['spec_rating'], bins=20, kde=True, color="orange")
plt.title("Spec Rating Distribution")
plt.xlabel("Spec Rating")
plt.ylabel("Count")
plt.show()
```

```python
# GPU Comparison
plt.figure(figsize=(10,15))
df_clean['GPU'].value_counts().head(10).plot(kind='barh')
plt.title('GPU Comparison')
plt.xlabel('Number of Laptops')
plt.ylabel('GPU')
plt.show()
```

```python
plt.figure(figsize=(10,5))
df_clean['brand'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title("Top 10 Laptop Brands by Count")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,6))
brand_price = df_clean.groupby("brand")["price"].mean().sort_values(ascending=False)[:10]
sns.barplot(x=brand_price.index, y=brand_price.values, palette="viridis")
plt.title("Average Price by Brand (Top 10)")
plt.xticks(rotation=45)
plt.ylabel("Average Price (INR)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Ram", y="price", data=df_clean, palette="Set2")
plt.title("Price Distribution by RAM Size")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))
top_processors = df_clean['processor'].value_counts().head(10).index
sns.boxplot(x="processor", y="spec_rating", 
            data=df_clean[df_clean['processor'].isin(top_processors)])
plt.title("Spec Rating by Processor (Top 10)")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8,6))
corr = df_clean[["price","spec_rating","display_size",
                 "resolution_width","resolution_height","warranty"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap (Numerical Features)")
plt.show()
```
